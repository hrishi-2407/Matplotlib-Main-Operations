# Line Chart with Customization

```python
years = [2016 + x for x in range(8)]
income = [55, 56 ,62, 61, 72, 72, 73, 75]
income_ticks = list(range(50, 80, 2))
```

```python
plt.plot(years, income)
plt.title('Income (YoY) in US Dollars', fontsize=20)
plt.xlabel('Year')
plt.ylabel('Yearly Income (in US Dollars)')
plt.yticks(income_ticks, [f'${x}k' for x in income_ticks])
```

```python
stock_a = [100,102,99,101,101,100,102]
stock_b = [90,95,102,104,105,103,109]
stock_c = [110,115,100,105,100,98,95]
day = ['Mon','Tues','Wed','Thurs','Fri','Sat','Sun']
```

```python
plt.plot(day, stock_a)
plt.plot(day, stock_b)
plt.plot(day, stock_c)
```

```python
plt.plot(day, stock_a, label='Google')
plt.plot(day, stock_b, label='Microsoft')
plt.plot(day, stock_c, label='Apple')
plt.legend(loc='lower center')
```

```python
# Four plots in one window
x = np.arange(0, 100)
fig, axes = plt.subplots(2,2)

axes[0,0].plot(x, np.sin(x))
axes[0,0].set_title('Sine wave')

axes[0,1].plot(x, np.cos(x))
axes[0,1].set_title('Cosine wave')

axes[1,0].plot(x, np.random.random(100))
axes[1,0].set_title('Random function')

axes[1,1].plot(x, np.log(x))
axes[1,1].set_title('Log function')

fig.suptitle('Four Plots')
plt.tight_layout()
```

```python
ax = plt.axes(projection='3d')

a = np.random.randint(0, 10, 100)
b = np.random.randint(0, 10, 100)
c = np.random.randint(0, 10, 100)

ax.scatter(a,b,c)
ax.set_title("3D Plot")
ax.set_xlabel('length')
ax.set_ylabel('breadth')
ax.set_zlabel('height')
```

```python
ax = plt.axes(projection='3d')

a = np.arange(0, 100, 0.1)
b = np.sin(a)
c = np.cos(a)

ax.plot(a,b,c)
```

```python
x = np.arange(0, 10, 0.1)
y = np.arange(0, 10, 0.1)

X, Y = np.meshgrid(x, y)

Z = np.sin(X) * np.cos(Y)

ax = plt.axes(projection = '3d')
ax.plot_surface(X, Y, Z)
```

```python
x = np.arange(0, 10, 0.1)
y = np.arange(0, 10, 0.1)

X, Y = np.meshgrid(x, y)

Z = np.sin(X) * np.cos(Y)

ax = plt.axes(projection = '3d')
ax.plot_surface(X, Y, Z, cmap='Spectral')
```
